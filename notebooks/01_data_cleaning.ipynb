# Module 0: Data Cleaning and Data Understanding

## Project Context
This notebook is the first step of the **Electric Vehicle Market Insights Analysis in Washington State** project.
The goal of this module is to understand the structure and quality of the dataset, clean the raw records, and prepare a reliable analytical dataset for downstream exploratory analysis, statistical modeling, and dashboard development.

## Objectives

- inspect the dataset structure, column types, and missing values
- identify potential data quality issues, such as duplicates, invalid entries, and inconsistent categories
- evaluate the usability of key variables, especially **State**, **Electric Range**, and **Base MSRP**
- standardize and simplify selected categorical fields for analysis
- create cleaned and analysis-ready features for later modules
