```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ev_data = pd.read_csv("C:/Users/15386/OneDrive - UBC/Downloads/EV_analysis/data/Electric_Vehicle_Population_Data.csv")
```

```python
# Keep the rows of WA only
ev_data['State'] = ev_data['State'].str.strip()
wa_data = ev_data[ev_data['State'] == 'WA'].copy()
```

```python
wa_data.info()
wa_data.isnull().sum()
```

```python
# Data check
wa_data['State'].value_counts(dropna=False)
```

```python
# Delete unused variables
wa_data = wa_data.drop(columns = [
    'VIN (1-10)',
    'Postal Code',
    'Legislative District',
    'DOL Vehicle ID',
    '2020 Census Tract'
])
```
